# matplotlib으로 간단한 차트 그려보기

### 간단한 Plot 차트 그리기

```python
import matplotlib.pyplot as plt
values = [10, 30, 20, 40, 30]
plt.plot(values)
```

### 간단한 Bar 차트 그리기

```python
items = ['A', 'B', 'C', 'D', 'E']
values = [10, 30, 20, 40, 30]
plt.bar(items, values)
```

### 간단한 Pie Chart 만들기

```python
items = ['A', 'B', 'C', 'D', 'E']
values = [10, 30, 20, 40, 30]

plt.pie(values, labels=items)
plt.show()
```

# 단계별로 옵션 배우기

### Step 1. 기본 차트 생성하기

```python
temp1994 = [39.6, 37.9, 37.9, 34.9, 33.9, 35.3, 35.9, 35.3, 32.5, 36.8]
temp2018 = [29.3, 33.9, 32.5, 32.4, 35.6, 35.9, 33.6, 33.8, 37.0, 29.3]

plt.plot(temp1994)
plt.plot(temp2018)

plt.show()
```

### Step 2. 범례 표시하기

```python
temp1994 = [39.6, 37.9, 37.9, 34.9, 33.9, 35.3, 35.9, 35.3, 32.5, 36.8]
temp2018 = [29.3, 33.9, 32.5, 32.4, 35.6, 35.9, 33.6, 33.8, 37.0, 29.3]

plt.plot(temp1994, label = '1994')
plt.plot(temp2018, label = '2018')
plt.legend()

plt.show()
```

### Step 3. 마커 적용하기

```python
temp1994 = [39.6, 37.9, 37.9, 34.9, 33.9, 35.3, 35.9, 35.3, 32.5, 36.8]
temp2018 = [29.3, 33.9, 32.5, 32.4, 35.6, 35.9, 33.6, 33.8, 37.0, 29.3]

plt.plot(temp1994, marker = 'o', label = '1994')  # marker: o, v, ^, s, +, .
plt.plot(temp2018, marker = 'o', label = '2018')
plt.legend()

plt.show()
```

### Step 4. Y축 범위 변경하기: y-axis limit

```python
temp1994 = [39.6, 37.9, 37.9, 34.9, 33.9, 35.3, 35.9, 35.3, 32.5, 36.8]
temp2018 = [29.3, 33.9, 32.5, 32.4, 35.6, 35.9, 33.6, 33.8, 37.0, 29.3]

plt.plot(temp1994, label = '1994', marker = 'o')
plt.plot(temp2018, label = '2018', marker = 'o')

plt.ylim(20, 45)

plt.legend()
plt.show()
```

### Step 5. X축 범위 변경하기: x-axis limit

```python
temp1994 = [39.6, 37.9, 37.9, 34.9, 33.9, 35.3, 35.9, 35.3, 32.5, 36.8]
temp2018 = [29.3, 33.9, 32.5, 32.4, 35.6, 35.9, 33.6, 33.8, 37.0, 29.3]

plt.plot(temp1994, label = '1994', marker = 'o')
plt.plot(temp2018, label = '2018', marker = 'o')

plt.ylim(20, 45)
plt.xlim(0, 9)

plt.legend()
plt.show()
```

### Step 6. 제목(Title) 표시하기

```python
temp1994 = [39.6, 37.9, 37.9, 34.9, 33.9, 35.3, 35.9, 35.3, 32.5, 36.8]
temp2018 = [29.3, 33.9, 32.5, 32.4, 35.6, 35.9, 33.6, 33.8, 37.0, 29.3]

plt.plot(temp1994, label = '1994', marker = 'o')
plt.plot(temp2018, label = '2018', marker = 'o')

plt.ylim(20, 45)
plt.xlim(2, 8)

plt.title("temperature")  # 한글로 할 경우 에러 발생

plt.legend()
plt.show()
```
